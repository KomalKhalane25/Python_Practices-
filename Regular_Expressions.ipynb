```python
test_phrase = 'sdsd..sssddd...sdddsddd...dsds...dsssss...sdddd'
pattern = 'sd' # 1 s followed by 1 d
re.findall(pattern,test_phrase)
```

```python
def multi_re_fimd(pattern,phrase):
  for x in pattern:
    print("Searching for {} in phrase {}".format(x,phrase))
    print(re.findall(x,phrase))
    print("\n")
```

```python
test_phrase = 'sdsd...ssssddd...sdddsddd...sddsd.ssddd.....sdsd..sdddds.sd.sd'
test_patterns = ['sd*',#s followed by 0 or more d
                 'sd+', # 1 s followed by 1 or more s
                 'sd?',# s followed by 0 or 1 d
                 'sd{3}',# s followed by 3 d
                 'sd{2,3}'] # s followed by 2,3 d
multi_re_fimd(test_patterns,test_phrase)
```

```python
test_phrase = 'sdsd...ssjssddd...sdddcsddd...sdhdsd.ssddd.....sdsd..sdddds.sd.sd'
test_patterns = ['[sd]', # either s or d
                 '[sd]+', # either s or d with concat
                 's[sd]', #s followed by either s or d
                 's[sd]+'] # s followed by either s or d with concat
multi_re_fimd(test_patterns,test_phrase)
```
