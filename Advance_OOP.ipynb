```python
class Animal:
  def __init__(self, name):
    self.name = name

  def speak(self):
    print(self.name)

class Dog(Animal):
  def speak(self):
    return self.name + " says woof!"
```

```python
fido = Dog("Sam")
fido.speak()
```

```python
a = Animal("fred")
a.speak()
```

There Are 5 man type of inheritance:
1. Single Inheritance
2. Multiple Inheritance
3. Multi-level Inheritance
4. hierarchical Inheritance
5. Hybrid Inheritance


```python
class Car:
  def __init__(self,wheels=4):
    self.wheels = wheels
```

```python
class Gosoline(Car):
  def __init__(self, engine="Gosolone",thnk_cap=20):
    Car.__init__(self)
    self.engine = engine
    self.thnk_cap = thnk_cap
    self.tank = 0 #represent corrent value in the tank

  def refuel(self):
    self.tank = self.thnk_cap
```

```python
class Electric(Car):
  def __init__(self, engine="Electric",kwh_cap=60):
    Car.__init__(self)
    self.engine = engine
    self.kwh_cap = kwh_cap
    self.kwh = 0 #represent corrent battery value

  def recharge(self):
    self.kwh = self.kwh_cap
```

```python
class Hybrid(Gosoline,Electric):
  def __init__(self,engine="Hybrid",thnk_cap=11,kwh_cap=20):
    self.engine = engine
    Gosoline.__init__(self,engine,thnk_cap)
    Electric.__init__(self,engine,kwh_cap)
```

```python
prius = Hybrid()
print(prius.engine)
```

```python
print(prius.kwh)
```

```python
prius.recharge()
print(prius.kwh)
```

```python
print(prius.tank)
```

```python
prius.refuel()
print(prius.tank)
```

```python
g = Gosoline()
print(g.tank)
```

```python
e = Electric()
print(e.kwh)
```

```python
e.recharge()
print(e.kwh)
```
