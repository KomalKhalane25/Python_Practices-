```python
s = "My name is komal"
count_vowel = 0
count_consonant = 0
for i in s:
    if i == 'a' or i=='e' or i=='i' or i=='o' or i=='u':
       count_vowel=count_vowel+1
    else:
        count_consonant=count_consonant+1
print("vowel",count_vowel)
print("consonant",count_consonant)
```

```python
lst = [1,2,3,4,5,6.1,4.2,5.,"komal","khalane","dhule"]
int_lst = []
float_lst = []
string_lst = []
for i in lst:
  if type(i)==int:
    int_lst.append(i)
  elif type(i)==float:
    float_lst.append(i)
  elif type(i)==str:
    string_lst.append(i)
print("int_lst",int_lst)
print("float_lst",float_lst)
print("string_lst",string_lst)

```
