```python
pwd = 123
count = 0
while count < 5:
  try:
    otp = int(input("Enter the otp :"))
    if otp == pwd:
      print("Transaction successfully")
      break
    else:
      print("Invalid otp , Please retray")
      count = count + 1
  except:
    print("Invalid otp , Please retray")
    count = count + 1
else:
  print("Transaction failed")

```

```python
def test(a):
  try:
    if a<4:
      b = a/(a-3)
    print("Value of b is",b)
  except Exception as e:
    print(e)
test(3)
```
