```python
print("Normal Dictionary")
d1 ={}
d1['a']="A"
d1['b']="B"

d2 = {}
d2['b'] = "B"
d2['a'] = "A"

d1 == d2
```

```python
from collections import OrderedDict
```

```python
d1 = OrderedDict()
d1['a']="A"
d1['b']="B"

d2 = OrderedDict()
d2['b'] = "B"
d2['a'] = "A"

d1 == d2
```

```python
from collections import defaultdict
```

```python
def default_value():
  return "Key does not exist"

d = defaultdict(default_value)
d['a'] = 1
d['b'] = 2

print(d["a"])
print(d["b"])
print(d["c"])
```
