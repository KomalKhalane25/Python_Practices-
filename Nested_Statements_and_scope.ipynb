```python
def calculator(num1,num2):
  if num1%2==0 and num2%2==0:
    return num1+num2
  elif num1%2!=0 and num2%2!=0:
    return abs(num1-num2)
  else:
    return num1*num2,num1/num2
```

```python
calculator(10,20)
```

```python
calculator(3,9)
```

```python
calculator(5,2)
```
