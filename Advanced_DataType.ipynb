```python
#Advanced List
def count_value(lst,num):
  count=0
  for i in lst:
    if i==num:
      count = count + 1
  return count
```

```python
count_value([1,1,1,2,4,5,1,3,5,5],5)
```

```python
count_value([11,44,55,77,11,22,11],11)
```
