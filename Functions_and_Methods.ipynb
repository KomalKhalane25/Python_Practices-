```python
def check_even_lst(x):
  odd_lst = []
  even_lst = []
  for i in x:
    if i%2==0:
      even_lst.append(i)
    else:
      odd_lst.append(i)
  return even_lst,odd_lst
```

```python
a,b = check_even_lst(lst)
print(a)
print(b)
```

```python
def lesser_of_two_evens(a,b):
    if a%2 == 0 and b%2 == 0:
        return min(a,b)
    else:
        return max(a,b)
```

```python
# Check
lesser_of_two_evens(2,4)
```
