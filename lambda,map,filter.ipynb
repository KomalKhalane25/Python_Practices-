```python
def check_even(num):
  if num%2==0:
    return "even"
  else:
    return "odd"
```

```python
check_even(5)
```

```python
check_even(2)
```

```python
check_even([1,2,5,6,])
```

```python
#filter() : are return all element output
def check_even(num):
  return num%2==0
list(map(check_even,[1,2,5,6,]))
```

```python

list(filter(check_even,[1,2,5,6,])) # filter are only return true value output
```

```python
#lambda : is an expression
```

lambda is an expression which allow us to create "anonymous" function. lambda is a single line of code not multiple block of code


```python
list(map(lambda num:num**2,[1,2,5,6,]))
```

```python
list(filter(lambda num:num%2==0,[1,2,5,6,3]))
```
